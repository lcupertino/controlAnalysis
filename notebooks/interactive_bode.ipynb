```python
def create_transfer_function(num, den):
    return control.tf(num, den)

def draw_Bode(gain: float, transfer_function: control.tf):
    G = gain*transfer_function
    return control.bode(G)
    
def draw_frame(gain, num, den):
    transfer_function = create_transfer_function(num, den)
    draw_Bode(gain, transfer_function)
```

```python
fig = plt.figure(figsize=(12,5))
```
